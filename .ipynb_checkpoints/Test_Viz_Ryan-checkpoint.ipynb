```python
import sqlite3 
import pandas as pd
```

```python
conn = sqlite3.connect("COVID.db")
```

```python
cursor = conn.cursor()
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cursor.fetchall())
```

```python
df = pd.read_sql_query("SELECT * FROM covid_data", conn)
df
```

```python
columns = pd.read_sql_query ("SELECT name FROM PRAGMA_TABLE_INFO('covid_data')", conn)
columns
```

1. How many countries are included in our data set?

```python

location = pd.read_sql_query("SELECT COUNT (DISTINCT location) FROM covid_data", conn)
location
```

2. When was the most recent entry?

```python
time = pd.read_sql_query("SELECT date FROM covid_data ORDER BY date DESC LIMIT 1", conn)
time
```

```python
tim = pd.read_sql_query("SELECT new_deaths_per_million FROM covid_data", conn)
tim
```

3. What country had the highest single day deaths per million?

```python
dea = pd.read_sql_query("SELECT location FROM covid_data WHERE new_deaths_per_million ORDER BY new_deaths_per_million ASC LIMIT 1", conn)
dea
```

4. Sélectionnez n'importe quel pays. Combien de fois le nombre de décès par million a-t-il dépassé 1 000 en une seule journée ? 

```python
test = pd.read_sql_query("SELECT location, COUNT(*) as NB FROM covid_data GROUP BY new_deaths_per_million ORDER BY NB asc", conn)
test
```

```python
deal = pd.read_sql_query("SELECT COUNT (*) FROM covid_data WHERE new_deaths_per_million > 1000", conn)
deal
```
