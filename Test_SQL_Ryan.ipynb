```python
df = pd.read_sql_query("SELECT * FROM covid_data", conn)
df
```

Requête pour afficher toutes les colonnes

```python
columns = pd.read_sql_query ("SELECT name FROM PRAGMA_TABLE_INFO('covid_data')", conn)
columns
```

1. How many countries are included in our data set?

```python
location = pd.read_sql_query("SELECT COUNT (DISTINCT location) FROM covid_data", conn)
location
```

2. When was the most recent entry?

```python
date = pd.read_sql_query("SELECT date FROM covid_data ORDER BY date DESC LIMIT 1", conn)
date
```

3. What country had the highest single day deaths per million?

```python
country_highest = pd.read_sql_query("SELECT location FROM covid_data WHERE new_deaths_per_million ORDER BY new_deaths_per_million ASC LIMIT 1", conn)
country_highest 
```

4. Select any country. How many times have the deaths per million exceeded 1,000 in one day? 

```python
#rand = pd.read_sql_query("SELECT location FROM covid_data GROUP BY RANDOM() LIMIT 1", conn)
#rand
```

```python
rand_1 = pd.read_sql_query("SELECT location, COUNT(*) FROM covid_data WHERE new_deaths_per_million > 1000 AND location = 'Poland'", conn)
rand_1
```

5. Extract the deaths per million for US, France, and India. Plot the time series data and compare results to what the media is currently saying

```python
france = pd.read_sql_query("SELECT date, new_deaths_per_million FROM covid_data WHERE location = 'France'", conn)
france
```

```python
india = pd.read_sql_query("SELECT date, new_deaths_per_million FROM covid_data WHERE location = 'India'", conn)
india
```

```python
us = pd.read_sql_query("SELECT date, new_deaths_per_million FROM covid_data WHERE location = 'United States'", conn)
us 
```

6. Plot the deaths per million and cases per million by continent for your birthday in 2020. What does the data say?

```python
continent = pd.read_sql_query("SELECT continent, COUNT (*) total_deaths_per_million FROM covid_data WHERE date = '2020-01-01' GROUP BY continent", conn)
continent
```

```python
continent_1 = pd.read_sql_query("SELECT continent, COUNT (*) total_cases_per_million FROM covid_data WHERE date = '2020-01-01' GROUP BY continent", conn)
continent_1
```
